```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('datasets/zomato.csv', encoding='utf8')
df=data.copy()
np_city=np.array(df['listed_in(city)'])
dic=dict()
for i in np_city:
    if i in dic.keys():
        dic[i]+=1
    else:
        dic[i]=1
x=[]
y=[]
for i in dic.keys():
    x.append(i)
    y.append(dic[i])
np_x=np.array(x)
np_y=np.array(y)
np_x=np_x[np.argsort(np_y)]
np_y=np.sort(np_y)
np_x=np_x[::-1]
np_y=np_y[::-1]
plt.scatter(np_x, np_y)
plt.xticks(rotation=90)
plt.xlabel('Cities', fontsize=16)
plt.ylabel('Occurance', fontsize=16)
plt.rcParams["figure.figsize"]=[15, 15]
plt.tick_params(labelsize=15)
plt.show()
for i in range(len(np_x)):
    print(np_x[i],'------>' , np_y[i])
```
