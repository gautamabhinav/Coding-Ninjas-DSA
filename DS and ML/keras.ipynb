```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
layer1=Dense(units=32, activation='tanh', input_dim=30)#input layer
model.add(layer1)
model.add(Dense(activation="relu", units=16))#hidden layer
model.add(Dense(activation="sigmoid", units=1))#output layer
```

```python
model.compile(metrics=['accuracy'], loss='binary_crossentropy', optimizer='adam')
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
data=load_breast_cancer()
x_train, x_test, y_train, y_test=train_test_split(data.data, data.target)
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=50, validation_data=(x_test, y_test))
```

```python
predictions=model.predict(x_test)
loss, accuracy=model.evaluate(x_test, y_test)
accuracy, loss
```
