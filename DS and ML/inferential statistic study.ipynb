```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('datasets/weight-height.csv')
```

```python
df
```

```python
samples=[]
for i in range(1, 1000):
    samples.append(df['Weight'].sample(30).mean())
```

```python
samples
```

```python
plt.hist(samples, bins=500, color='r')
plt.show()
```

```python
#now if we increase the sample size,
samples=[]
for i in range(1, 1000):
    samples.append(df['Weight'].sample(5000).mean())
plt.hist(samples, bins=500, color='r')
plt.show()
```
